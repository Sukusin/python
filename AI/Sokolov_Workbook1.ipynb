```python
#4.3.5
import math as m



while True:
	
	do = input('"break для остановки"'+ '\n' +"Действие(+, -, /, *, e^(x + y), sin(x + y), cos(x + y), x^y):")

	if do == 'break':
		break

	x = float(input("Первое число(x):"))
	y = float(input("Второе число(y):"))
	c = 0.0
	
	if do == '+':
		c = x + y
		print('Результат: ' + str(c))

	if do == '-':
		c = x - y
		print('Результат: ' + str(c))

	if do == '*':
		c = x * y
		print('Результат: ' + str(c))

	if do == '/':
		c = x + y
		print('Результат: ' + str(c))

	if do == 'e^(x + y)':
		c = m.e ** (x + y)
		print('Результат: ' + str(c))

	if do == 'sin(x + y)':
		c = m.sin(x + y)
		print('Результат: ' + str(c))

	if do == 'cos(x + y)':
		c = m.cos(x + y)
		print('Результат: ' + str(c))

	if do == 'x^y':
		c = x ** y
		print('Результат: ' + str(c))
	
```
